```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(df, target, test_size = 0.25, random_state = 0)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test= sc.transform(x_test)
```

```python
from sklearn.linear_model import LinearRegression,Ridge,Lasso,ElasticNet

lr=LinearRegression()
lr.fit(x_train,y_train)
pred=lr.predict(x_test)
rid = Ridge()
rid.fit(x_train,y_train)
rid_pred = rid.predict(x_test)
las = Lasso()
las.fit(x_train,y_train)
las_pred = las.predict(x_test)
en = ElasticNet()
en.fit(x_train,y_train)
en_pred = en.predict(x_test)
from sklearn.svm import SVR
svr = SVR()
svr.fit(x_train,y_train)
svr_pred = svr.predict(x_test)
from sklearn.tree import DecisionTreeRegressor
dtr = DecisionTreeRegressor()
dtr.fit(x_train,y_train)
dtr_pred = dtr.predict(x_test)
from sklearn.ensemble import RandomForestRegressor
rfr = RandomForestRegressor()
rfr.fit(x_train,y_train)
rfr_pred = rfr.predict(x_test)
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
print("LinearRegression")
print((mean_squared_error,y_test,pred))
print(mean_absolute_error(y_test,pred))
print(r2_score(y_test,pred))
print("-----------------------")
print("Lasso")
print(mean_squared_error(y_test,las_pred))
print(mean_absolute_error(y_test,las_pred))
print(r2_score(y_test,las_pred))
print("-----------------------")
print("ridge")
print(mean_squared_error(y_test,rid_pred))
print(mean_absolute_error(y_test,rid_pred))
print(r2_score(y_test,rid_pred))
print("-----------------------")
print("ElasticNet")
print(mean_squared_error(y_test,en_pred))
print(mean_absolute_error(y_test,en_pred))
print(r2_score(y_test,en_pred))
print("-----------------------")
print("support vector regressor")
print(mean_squared_error(y_test,svr_pred))
print(mean_absolute_error(y_test,svr_pred))
print(r2_score(y_test,svr_pred))
print("-----------------------")
print("Decision tree regressor")
print(mean_squared_error(y_test,dtr_pred))
print(mean_absolute_error(y_test,dtr_pred))
print(r2_score(y_test,dtr_pred))
print("-----------------------")
print("random forest regression")
print(mean_squared_error(y_test,rfr_pred))
print(mean_absolute_error(y_test,rfr_pred))
print(r2_score(y_test,rfr_pred))
print("-----------------------")
```
